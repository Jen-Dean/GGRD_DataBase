```python
import pandas as pd
import openpyxl
```

```python
#Pulling the Excel File & Sheet
path = "STATS/all_stars_2018/test.xlsx"
stats_df = pd.read_excel(path, sheet_name="Lineups", engine="openpyxl")
```

```python
#Check that the data came through
stats_df.head()
```

```python
stats_df = stats_df.fillna(0)
stats_df = stats_df.drop([0], axis = "index")
stats_df = stats_df.reset_index()
stats_df.head()
```

```python
stats_df = stats_df.rename(columns=stats_df.iloc[0]).drop(stats_df.index[0])
stats_df = stats_df.drop([1, "noPivot", "Box", 0.0, 0, "Team Roster"], axis = 1)
stats_df.head()
```

```python
stats_df = stats_df.rename(columns={stats_df.columns[0]: "jam_number",
                                   stats_df.columns[1]: "home_jammer_number",
                                   stats_df.columns[2]: "home_pivot_number",
                                   stats_df.columns[3]: "home_blocker_1_number",
                                   stats_df.columns[4]: "home_blocker_2_number",
                                   stats_df.columns[5]: "home_blocker_3_number",
                                   stats_df.columns[6]: "redundant",
                                   stats_df.columns[7]: "away_jammer_number",
                                   stats_df.columns[2]: "away_pivot_number",
                                   stats_df.columns[3]: "away_blocker_1_number",
                                   stats_df.columns[4]: "away_blocker_2_number",
                                   stats_df.columns[5]: "away_blocker_3_number"})
stats_df = stats_df.drop(["redundant"], axis = 1)
stats_df
```

```python
stats_df = stats_df.dropna(axis=1, how="all")
stats_df = stats_df.drop([0], axis = "index")
stats_df = stats_df.dropna(axis="index", how="all").reset_index()
stats_df = stats_df.rename(columns=stats_df.iloc[0]).drop(stats_df.index[0])
stats_df
```
